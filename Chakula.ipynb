```python
def sel_res_user_xi(user_rest_data,i):
    X = user_rest_data.iloc[i]
    li_zero = []
    li_non_zero = []
    for id in range(1, len(X)):
        if(X.iloc[id] == 0):
            li_zero.append(id-1)
        else:
            li_non_zero.append(id-1)
    return li_zero, li_non_zero
```

```python
print(user_rest_data.head())
li0, li1 = sel_res_user_xi(user_rest_data, 3)
print(li1)
```

```python
def dataframe_selector(rest_data, li_zero):
    X = rest_data.copy()
    for itr in li_zero:
        X.drop(itr, inplace=True)
    return X
```

```python
li0, li1 = sel_res_user_xi(user_rest_data, 3)
print(len(li0), len(li1))
sd = dataframe_selector(rest_data, li0)
print(sd.shape)
print(li1)
print(sd.tail())
```

```python
def k_nearest(df, rest_id, k = 30):
    neigh = KNN(n_neighbors = k, metric='cosine').fit(df.iloc[:,1:])
    distances, indices = neigh.kneighbors([df.iloc[rest_id,1:]])
    distances = 1 - distances 
    return indices, distances
    # cosine = sklearn.metrics.pairwise.cosine_similarity([df.iloc[rest_id, 1:]], df.iloc[:,1:]) 
    # cosine[0].sort()
    # return cosine[0][0:k]
```
